```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

font = {'family' : 'normal',
        'weight' : 'normal',
        'size'   : 22.5}

matplotlib.rc('font', **font)
filename = 'HP_10000.csv'
df = pd.read_csv('HP_10000.csv')
#df1= pd.read_csv('FEC_10000.csv')
df.columns#, df1.columns
```

```python
#Plot parameters

xlow = -1.5
xhigh = 2.5


x=df['Target']
y=df['Prediction']
y_err=[df['LE'],df['UE']]

#x1=df1['Target']
#y1=df1['Prediction']
#y1_err=[df1['LE'],df1['UE']]
```

```python
def scatplot(x_data, y_data, error_data, x_label, y_label, title):
    _, ax = plt.subplots(figsize=(12,12))
    # Draw bars, position them in the center of the tick mark on the x-axis
    ax.scatter(x_data, y_data, color = 'red',s=30)
    # Draw error bars to show standard deviation, set ls to 'none'
    # to remove line between points
    ax.errorbar(x_data, y_data,  
                yerr = error_data, 
                color = 'black', 
                ls = 'none', 
                lw = 0.5, capsize = 5)
    plt.xlim(xlow,xhigh)
    plt.ylim(xlow,xhigh)
    ax.plot([xlow, xhigh], [xlow, xhigh], color='gray', linestyle='-', linewidth=0.5)
    ax.set_ylabel(y_label)
    ax.set_xlabel(x_label)
    ax.set_title(title,fontsize=22.5)
    

# Call the function to create plot
scatplot(x_data = x, 
         y_data = y,
         error_data = y_err,
         x_label = 'Target',
         y_label = 'Prediction',
         title = 'HP, B=10000')

plt.annotate('Slope = 0.879', 
             xy=(-1.25, 2.0),  
             xycoords='data',
             textcoords='offset points',
             fontsize='16'
            # arrowprops=dict(arrowstyle="->")
            )
plt.annotate('Mean Squared Error = 0.0153', 
             xy=(-1.25, 1.8),  
             xycoords='data',
             textcoords='offset points',
             fontsize='16'
            # arrowprops=dict(arrowstyle="->")
            )
#Plotting prediction for inputs without 
#plt.scatter(x1, y1, color = 'green',s=30)
#plt.errorbar(x1, y1,yerr = y1_err, color = 'k', ls = 'none', lw = 0.5, capsize = 5)
#plt.show()
```
