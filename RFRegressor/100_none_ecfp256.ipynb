```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

font = {'family' : 'normal',
        'weight' : 'normal',
        'size'   : 22.5}

matplotlib.rc('font', **font)
filename = '100_none_ecfp256.csv'
df = pd.read_csv(filename)
#df1= pd.read_csv('FEC_10000.csv')
df.columns#, df1.columns
```

```python
#Plot parameters

xlow = -1.5
xhigh = 2.5


x=df['Target']
y=df['Prediction']
y_err=[df['LE'],df['UE']]

#x1=df1['Target']
#y1=df1['Prediction']
#y1_err=[df1['LE'],df1['UE']]
```

```python
def scatplot(x_data, y_data, error_data, x_label, y_label, title):
    _, ax = plt.subplots(figsize=(12,12))
    # Draw bars, position them in the center of the tick mark on the x-axis
    ax.scatter(x_data, y_data, color = 'red',s=30)
    # Draw error bars to show standard deviation, set ls to 'none'
    # to remove line between points
    ax.errorbar(x_data, y_data,  
                yerr = error_data, 
                color = 'black', 
                ls = 'none', 
                lw = 0.5, capsize = 5)
    plt.xlim(xlow,xhigh)
    plt.ylim(xlow,xhigh)
    ax.plot([xlow, xhigh], [xlow, xhigh], color='gray', linestyle='-', linewidth=0.5)
    ax.set_ylabel(y_label)
    ax.set_xlabel(x_label)
    ax.set_title(title,fontsize=22.5)
    

# Call the function to create plot
scatplot(x_data = x, 
         y_data = y,
         error_data = y_err,
         x_label = 'Target',
         y_label = 'Prediction',
         title = 'RF_100_none_ecfp256, B=100')

plt.annotate('Slope = 0.940', 
             xy=(-1.25, 1.7),  
             xycoords='data',
             textcoords='offset points',
             fontsize='16'
            # arrowprops=dict(arrowstyle="->")
            )
plt.annotate('Mean Squared Error = 0.0262', 
             xy=(-1.25, 1.5),  
             xycoords='data',
             textcoords='offset points',
             fontsize='16'
            # arrowprops=dict(arrowstyle="->")
            )
plt.annotate('N---N+-p', 
             xy=(1.91, 1.7),  
             xycoords='data',
             textcoords='offset points',
             fontsize='16'
            # arrowprops=dict(arrowstyle="->")
            )
plt.annotate('N---N+-m', 
             xy=(1.76, 1.78),  
             xycoords='data',
             textcoords='offset points',
             fontsize='16'
            # arrowprops=dict(arrowstyle="->")
            )
plt.annotate('S--p', 
             xy=(-1.21, -0.5),  
             xycoords='data',
             textcoords='offset points',
             fontsize='16'
            # arrowprops=dict(arrowstyle="->")
            )
#plt.annotate('N(Me)2-p', 
            # xy=(-0.83, -0.63),  
             #xycoords='data',
            # textcoords='offset points',
            # fontsize='16'
            # arrowprops=dict(arrowstyle="->")
            #)
plt.annotate('S--m', 
             xy=(-0.36, -0.41),  
             xycoords='data',
             textcoords='offset points',
             fontsize='16'
            # arrowprops=dict(arrowstyle="->")
            )            
#Plotting prediction for inputs without 
#plt.scatter(x1, y1, color = 'green',s=30)
#plt.errorbar(x1, y1,yerr = y1_err, color = 'k', ls = 'none', lw = 0.5, capsize = 5)
#plt.show()
```
